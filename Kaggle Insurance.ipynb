```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import statsmodels.api as sm 
```

```python
read_path = r'D:\Insurance Kaggle\\'
read_file = 'insurance.csv'
df = pd.read_csv(read_path+read_file)
df.head()
```

```python
#turning regions into categorical variable

regions = df['region'].unique().tolist()
regions

for i in range(0, len(regions)): 
    df[regions[i]] =  0
    mask = df['region'] ==regions[i]
    df.loc[mask, regions[i]] = 1
```

```python
mask = df['sex'] == 'male'
df.loc[mask, 'sex'] = 1
df.loc[~mask, 'sex'] = 0
```

```python
mask = df['smoker'] == 'yes'
df.loc[mask, 'smoker'] = 1
df.loc[~mask, 'smoker'] = 0 
df
```

```python
reg = df[['age','sex','children','smoker','charges','southwest','southeast','northwest','northeast']]
reg
```

```python
reg.describe()
```

```python
cols = reg.columns.tolist()
for i in range(0, len(cols)):
    print(reg[cols[i]].unique())
```

```python
reg['sex'] = pd.to_numeric(reg['sex'])
reg['smoker'] = pd.to_numeric(reg['smoker'])
reg.dtypes
```

```python
def regression(df, x, y): 
    Y =  df[y].values
    X = df[x].values
    model = sm.OLS(Y,X)
    res = model.fit()
    return(res)

r = regression(reg.dropna(), ['age','sex','children','smoker','southwest','southeast','northwest','northeast'], 'charges')
r.summary()
```
